```python
import sbol2
```

```python
from buildcompiler.sbol2build import rebase_restriction_enzyme, add_object_to_doc

sbol2.Config.setHomespace("https://SBOL2Build.org")
sbol2.Config.setOption(sbol2.ConfigOptions.SBOL_COMPLIANT_URIS, True)
sbol2.Config.setOption(sbol2.ConfigOptions.SBOL_TYPED_URIS, False)

protein_doc = sbol2.Document()


BsaI_def = rebase_restriction_enzyme("BsaI")

RE_sourcing = sbol2.Activity("restriction_enzyme_purchase")
RE_sourcing.name = "Restriction Enzyme Purchase"

BsaI_imp = sbol2.Implementation(f"{BsaI_def.displayId}_impl")

BsaI_imp.built = BsaI_def.identity
BsaI_imp.wasGeneratedBy = RE_sourcing.identity

BbsI_def = rebase_restriction_enzyme("BbsI")

BbsI_imp = sbol2.Implementation(f"{BbsI_def.displayId}_impl")

BbsI_imp.built = BbsI_def.identity
BbsI_imp.wasGeneratedBy = RE_sourcing.identity


ligase = sbol2.ComponentDefinition("T4_Ligase")
ligase.name = "T4_Ligase"
ligase.types = [sbol2.BIOPAX_PROTEIN]
ligase.roles = ["http://identifiers.org/ncit/NCIT:C16796"]

ligase_sourcing = sbol2.Activity("ligase_purchase")
ligase_sourcing.name = "Ligase Purchase"

T4_imp = sbol2.Implementation(f"{ligase.displayId}_impl")
T4_imp.built = ligase.identity
T4_imp.wasGeneratedBy = ligase_sourcing.identity

add_object_to_doc(ligase, protein_doc)
add_object_to_doc(T4_imp, protein_doc)
add_object_to_doc(BsaI_def, protein_doc)
add_object_to_doc(BsaI_imp, protein_doc)
add_object_to_doc(BbsI_def, protein_doc)
add_object_to_doc(BbsI_imp, protein_doc)
add_object_to_doc(RE_sourcing, protein_doc)

protein_doc.write("proteins.xml")
```
